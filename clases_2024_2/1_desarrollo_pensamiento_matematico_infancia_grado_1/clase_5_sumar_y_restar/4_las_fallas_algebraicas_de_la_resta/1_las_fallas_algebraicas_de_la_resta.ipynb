## El componente histórico y epistemológico    

La **resta** en los números naturales es una de las operaciones fundamentales en matemáticas, y su enseñanza ha evolucionado a lo largo de la historia. A continuación, se presentan elementos históricos y epistemológicos sobre la **resta** en los números naturales, así como enfoques para su enseñanza mediante el **ábaco**, **cuadernos Jupyter** y el **lenguaje Python**.

### 1. Elementos históricos de la resta:
#### Orígenes y evolución:
- **Antigua Mesopotamia**: Las primeras civilizaciones, como la mesopotámica, ya utilizaban formas primitivas de sustracción en su sistema numérico sexagesimal (base 60). Los escribas aprendían a restar para resolver problemas de repartición de bienes o medir tierras.
- **Antiguo Egipto**: Los egipcios también desarrollaron técnicas de sustracción utilizando jeroglíficos numéricos. Aunque su enfoque primario era la adición, la resta se comprendía como el proceso inverso.
- **Matemáticos griegos**: La noción de resta, como operación inversa de la suma, fue reconocida por matemáticos griegos, aunque su enfoque principal estaba en la geometría.
- **Edad Media y Renacimiento**: Con la introducción de los números arábigos y el uso del cero, la resta se formalizó como una operación aritmética sistemática en Europa, facilitada por el uso de algoritmos más eficientes.



#### La resta como operación inversa:  

Históricamente, la resta se entendió como la **operación inversa de la suma**. Esto se observa en la notación algebraica moderna, donde:
$$
a - b = c \quad \text{si y sólo si} \quad a = b + c
$$
La enseñanza de la resta como una operación inversa ayuda a los estudiantes a conectarla con la suma y facilita la comprensión de problemas de búsqueda de diferencias o cambios.



### 2. Elementos epistemológicos de la resta en números naturales:
La **epistemología** de la resta, en términos de su comprensión en la historia de las matemáticas y su enseñanza, gira en torno a los siguientes puntos:

- **Descomposición y complementos**: Uno de los primeros enfoques fue ver la resta como un proceso de descomposición o complementación. Por ejemplo, si se tenía un número 7 y se necesitaba restar 4, la comprensión inicial era descomponer 7 en 4 y un "resto" de 3.
- **Restar como proceso de comparación**: Epistemológicamente, la resta puede interpretarse como una forma de **comparación** entre dos cantidades, preguntando "¿cuánto falta?" o "¿cuánto sobra?".
- **Concepto de número cero**: La introducción del **cero** fue clave para entender la resta en su totalidad. Sin el cero, ciertas restas no podían ser resueltas fácilmente en sistemas numéricos antiguos, ya que la idea de “nada” o “ningún valor” no estaba formalizada.
- **La propiedad no conmutativa**: A diferencia de la suma, la resta no es conmutativa, lo cual es un aspecto importante en su enseñanza. Es decir, $ a - b \neq b - a $, excepto cuando $ a = b $.
- **Resta con números naturales**: En los números naturales, la resta es una operación que, por definición, no da como resultado números negativos. Esta limitación epistemológica fue superada más adelante con la introducción de los números enteros.

### 3. Enseñanza de la resta con ábaco:
El **ábaco** es una herramienta histórica utilizada desde hace milenios para la enseñanza de las operaciones aritméticas. En el contexto de la resta, el ábaco permite una representación visual y manipulativa de la operación, lo cual es particularmente útil para los niños que están aprendiendo los conceptos básicos de las matemáticas.

- **Modelo manipulativo**: Usar el ábaco para la resta ayuda a los estudiantes a visualizar cómo una cantidad disminuye. Por ejemplo, para resolver $ 8 - 5 $, el estudiante coloca 8 bolitas en el ábaco y luego retira 5, quedando con el resultado de 3.
- **Descomposición visual**: La operación de quitar físicamente las bolitas del ábaco refuerza la idea de descomponer una cantidad en partes, lo que es útil para la comprensión profunda de la resta.
- **Desarrollo del pensamiento lógico**: La manipulación del ábaco para restar promueve el razonamiento lógico en los estudiantes, ya que pueden explorar y verificar sus respuestas de manera tangible.

- **Integración con gráficos**: Python también permite utilizar bibliotecas como **matplotlib** para visualizar operaciones de resta de manera gráfica, lo que puede ayudar a los estudiantes a relacionar conceptos algebraicos con representaciones visuales.

### Ejemplo de código en cuadernos Jupyter para enseñar la resta:

````python

import matplotlib.pyplot as plt

# Definir los números iniciales y las cantidades a restar
iniciales = [20, 15, 10, 5]
restas = [5, 3, 2, 1]
resultados = [inicial - resta for inicial, resta in zip(iniciales, restas)]

# Visualizar el proceso de resta
plt.bar(range(len(iniciales)), iniciales, label='Valores iniciales')
plt.bar(range(len(restas)), restas, label='Valores restados', alpha=0.7)
plt.bar(range(len(resultados)), resultados, label='Resultados finales', alpha=0.4)
plt.legend()
plt.show()
````

1. **Propiedad Conmutativa (No se cumple):**  
   Los estudiantes ejecutarán el siguiente código en un Jupyter Notebook para verificar que la propiedad conmutativa no se cumple en la resta:
   


````python

# Verificación de la propiedad conmutativa
def resta_conmutativa(a, b):
    return a - b == b - a

a = 8
b = 3
print(f"¿La resta es conmutativa? {resta_conmutativa(a, b)}")

````

   **Resultado esperado:** `False`, ya que \(8 - 3 \neq 3 - 8\).

2. **Propiedad Asociativa (No se cumple):**  
   Luego, los estudiantes probarán si la resta es asociativa:
   

````python
   
   # Verificación de la propiedad asociativa
   def resta_asociativa(a, b, c):
       return (a - b) - c == a - (b - c)

   a = 10
   b = 5
   c = 2
   print(f"¿La resta es asociativa? {resta_asociativa(a, b, c)}")
   
````

````python
   **Resultado esperado:** `False`, ya que \((10 - 5) - 2 \neq 10 - (5 - 2)\).
````

3. **Elemento Neutro (No se cumple):**  
   Verificar si existe un elemento neutro para la resta:
   
   ```python
   # Verificación del elemento neutro
   def resta_neutro(a):
       return a - 0 == a

   a = 7
   print(f"¿Existe un elemento neutro para la resta? {resta_neutro(a)}")
   ```
   **Resultado esperado:** La resta tiene un "cuasi-elemento neutro" con el cero, pero no un verdadero neutro, ya que \(a - 0 = a\), pero \(0 - a \neq 0\).

4. **Distribución de la resta (No se cumple):**  
   Probar la propiedad distributiva de la resta frente a la suma:
   

````python
# Verificación de la propiedad distributiva
def resta_distributiva(a, b, c):
    return (a - (b + c)) == ((a - b) + (a - c))

a = 9
b = 3
c = 2
print(f"¿La resta es distributiva? {resta_distributiva(a, b, c)}")
   
````

   **Resultado esperado:** `False`, ya que la resta no distribuye sobre la suma.
