````python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/desarrollo_pensamiento_matematico_infancia.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) # hacer que todos los valores de la columna Parcial seen de tipo float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()
    guardar_cambios()


````

### **1. Introducción al Sistema Numérico Posicional (10 minutos)**

**Objetivo:** Comprender el sistema numérico posicional en base 10 y cómo se utiliza para representar números del 1 al 100.

- **Explicación:** Comienza explicando que el sistema numérico que usamos es posicional, lo que significa que el valor de un dígito depende de su posición dentro del número.
- **Ejemplo:** Usa números como 23 y 87 para ilustrar cómo la posición de cada dígito (unidades y decenas) determina su valor.
- **Actividad:** Pide a los estudiantes que escriban varios números entre 1 y 100 y que identifiquen el valor posicional de cada dígito en cada número.

**Propósito:** Establecer la base para comprender cómo los números naturales se representan y cómo se realiza la suma en este sistema.


Aquí tienes un programa en Python que ilustra cómo la posición de cada dígito en un número, como 23 y 87, determina su valor. Este programa descompone el número en unidades y decenas y muestra el valor correspondiente de cada dígito según su posición.


````python
def descomponer_numero(numero):
    decenas = numero // 10  # Obtener el dígito de las decenas
    unidades = numero % 10   # Obtener el dígito de las unidades
    
    print(f"El número {numero} se descompone en:")
    print(f"{decenas} decenas, lo que equivale a {decenas * 10}")
    print(f"{unidades} unidades, lo que equivale a {unidades}")
    print(f"Sumando {decenas * 10} + {unidades} = {numero}")
    print("")

# Ejemplos con los números 23 y 87
descomponer_numero(23)
descomponer_numero(87)

````


### Explicación del código:
- **descomponer_numero(numero):** Esta función toma un número y lo descompone en sus componentes de decenas y unidades.
- **decenas = numero // 10:** Divide el número por 10 para obtener el dígito de las decenas.
- **unidades = numero % 10:** Utiliza el operador módulo (%) para obtener el dígito de las unidades.
- **El programa imprime cómo se descompone el número, mostrando la cantidad de decenas y unidades, así como el valor total.**

### Resultado esperado:
Al ejecutar el programa, se mostrará lo siguiente:



```

El número 87 se descompone en:
8 decenas, lo que equivale a 80
7 unidades, lo que equivale a 7
Sumando 80 + 7 = 87
```



Este programa ilustra cómo cada dígito en un número contribuye a su valor total según su posición (decenas o unidades).


### **2. Exploración de la Suma en los Naturales (15 minutos)**

**Objetivo:** Realizar sumas sencillas de números del 1 al 100, reconociendo patrones y observando la relación con el sistema posicional.

- **Explicación:** Introduce la suma como la operación de combinar cantidades y enfatiza que al sumar dos números naturales, el resultado también es un número natural.
- **Propiedades de la Suma:**
  - **Propiedad Conmutativa:** A + B = B + A. Explica que el orden de los sumandos no afecta el resultado.
  - **Propiedad Asociativa:** (A + B) + C = A + (B + C). Muestra cómo agrupar sumandos de manera diferente no cambia la suma.
  - **Elemento Neutro:** A + 0 = A. Resalta que sumar cero a un número no cambia su valor.
- **Actividad:** Proporciona ejercicios donde los estudiantes sumen pares de números (por ejemplo, 23 + 45, 56 + 29) y verifiquen las propiedades con ejemplos.

**Propósito:** Ayudar a los estudiantes a internalizar las propiedades fundamentales de la suma en el contexto de los números naturales.
